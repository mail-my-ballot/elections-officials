## Local Data

## These are suffixes used for cities

```python
df_local['jurisdictionName'].str.split(' ').apply(lambda x: x[-1]).value_counts()
```

```python
def extract_address(lines):
    addr_lines = []
    start_address = False
    for line in lines:
        if line == '' and not start_address:
            start_address = True
            continue
            
        if line == '':
            break
        
        if start_address:
            addr_lines += [line]
        
    return ', '.join(addr_lines)
        

df_local['address'] = df_local['info'].apply(extract_address)

df_local['address'].sample(10, random_state=42)
```

```python
def get_michigan_index(parts):
    for i, p in enumerate(parts):
        if p.startswith('Michigan'):
            return i
        
    return None

def get_city(address):
    parts = address.split(', ')
    michigan_idx = get_michigan_index(parts)
    return parts[michigan_idx - 1]

df_local['addr_city'] = df_local['address'].apply(get_city)
df_local['addr_city'].sample(10, random_state=42)
```

```python
df_local['within'] = df_local.apply(lambda x: x['addr_city'] in x['jurisdictionName'], axis=1)
print("Only about {} are within their own townships ".format(df_local['within'].mean()))
```

```python
df_local.loc[~df_local['within'], ['addr_city', 'jurisdictionName', 'CountyName']].sample(20, random_state=42)
```

```python
df_local['maybe_within'] = df_local['addr_city'].apply(
    lambda x: df_local['jurisdictionName'].str.startswith(x).any()
)
print("Only about {} are possibly within an existing township ".format(df_local['maybe_within'].mean()))
```

```python
df_local.loc[~df_local['maybe_within'], ['addr_city', 'jurisdictionName', 'CountyName']].sample(20, random_state=42)
```
